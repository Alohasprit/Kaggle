# DT Leve 3 Day-2 Titanic Decision Tree

## 1. Import Requred Libraries

```python
import pandas as pd
```

## 2. Load Data

```python
train = pd.read_csv("./data/train.csv", index_col = "PassengerId")
test = pd.read_csv("./data/test.csv", index_col='PassengerId')
```

```python
train.head()
```

## 3. Data Preprocessing

```python
train.loc[train["Sex"]=='male', "Sex_encode"]=0
train.loc[train["Sex"]=='female', "Sex_encode"]=1
test.loc[test["Sex"]=='male', "Sex_encode"]=0
test.loc[test["Sex"]=='female', "Sex_encode"]=1
```

```python
train.head()
```

```python
test.head()
```

```python
train.head()
```

```python
train['Fare_fillin'] = train["Fare"]
test['Fare_fillin'] = test["Fare"]
```

```python
train.loc[train['Fare'].isnull(), 'Fare_fillin']=0
test.loc[test['Fare'].isnull(), 'Fare_fillin']=0
```

```python
train['Embarked_C'] = (train['Embarked']=='C')
train['Embarked_S'] = (train['Embarked']=='S')
train['Embarked_Q'] = (train['Embarked']=='Q')
```

```python
test['Embarked_C'] = (test['Embarked']=='C')
test['Embarked_S'] = (test['Embarked']=='S')
test['Embarked_Q'] = (test['Embarked']=='Q')
```

## 4. Feature Engineering

```python
feature_name = ['Pclass', 'Sex_encode', "Fare_fillin", 'Embarked_C', 'Embarked_S', 'Embarked_Q']
feature_name
```

```python
print(train[feature_name].shape)
train[feature_name].head()
```

```python
print(test[feature_name].shape)
test[feature_name].head()
```

```python
label_name = "Survived"
y_train = train[label_name]
y_train, y_train.shape
```

## 5. Training Using Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth=5)
model
```

```python
x_train = train[feature_name]
```

```python
model.fit(x_train, y_train)
```

```python
x_test = test[feature_name]
```

```python
predictions = model.predict(x_test)
```

```python
predictions[:10]
```

```python
predictions.shape
```

## 6. Submit

```python
submit = pd.read_csv("./data/gender_submission.csv", index_col ="PassengerId")
```

```python
submit["Survived"] = predictions
submit.head()
```

```python
submit.to_csv("./data/decision-tree.csv")
```
