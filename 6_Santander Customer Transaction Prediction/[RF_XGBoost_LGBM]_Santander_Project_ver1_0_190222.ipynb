```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```
train = pd.read_csv('./train.csv', index_col='ID_code')
test = pd.read_csv('./test.csv', index_col='ID_code')
```

```
print(train.shape)
train.head()
```

```
print(test.shape)
test.head()
```

```
y_train = train['target']
print(y_train.shape)
y_train.head()
```

```
x_train = train.loc[:, train.columns != 'target']
print(x_train.shape)
x_train.head()
```

```
x_test = test
print(x_test.shape)
x_test.head()
```

### 1. RF with Cross Validation

```
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

max_depth_list = range(50,70)
score =[]

for max_depth in max_depth_list:
    model_1= RandomForestClassifier(random_state=37, max_depth=max_depth)
    val_score = cross_val_score(model_1, x_train, y_train, cv=5, scoring ='roc_auc').mean()
    score.append({
                'max_depth': max_depth,
                'roc_auc' : val_score
    })

score = pd.DataFrame(score)

plt.plot(data=score, x='max_depth', y='roc_auc', label = 'val_score')
plt.legend()
```

### 2. XGBoost with Cross Validation

```
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score

max_depth_list_2 = range(50,70)
score_2 =[]

for max_depth in max_depth_list_2:
    model_2 = XGBClassifier(random_state= 37, n_jobs= -1, max_depth= max_depth)
    val_score = cross_val_score(model_2, x_train, y_train, cv=5, scoring ='roc_auc').mean()
    
    score_2.append({
        'max_depth' : max_depth,
        'roc_auc' : val_score
    })

score_2 = pd.DataFrame(score_2)

plt.plot(data=score_2, x='max_depth', y='roc_auc', label='val_score')
plt.legend()
```

### 3. LGBM with Cross Validation

```
from lightgbm import LGBMClassifier
from sklearn.model_selection import cross_val_score

max_depth_list_3 = range(50,70)
score_3 =[]

for max_depth in max_depth_list_3:
    model_3 =  LGBMClassifier(n_estimators = 15,random_state = 37, max_depth=max_depth)
    val_score = cross_val_score(model_3, x_train, y_train, cv=5, scoring ='roc_auc').mean()
    
    score_3.append({
        'max_depth' : max_depth,
        'roc_auc' : val_score
    })
    
score_3 = pd.DataFrame(score_3)

plt.plot(data=score_3, x='max_depth', y='roc_auc', label='val_score')
plt.legend()
```
