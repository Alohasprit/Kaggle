```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
```

```python
df = pd.read_csv('../../data/creditcard.csv')
print(df.shape)
df.head()
```

```python
df.isnull().sum().max()
```

```python
df.dtypes.value_counts()
```

```python
df.columns
```

```python
round(df['Class'].value_counts()[0]/len(df)*100, 2)
```

```python
#integer 값의 countplot 

colors = ["#0101DF", "#DF0101"]

sns.countplot('Class', data=df, palette=colors)
plt.title("Class Distributions \n (0:No Fraud, 1:Fraud)", fontsize=14)
```

```python
# subplot & float distplot

fig, ax = plt.subplots(1,2, figsize=(18,4))

sns.distplot(df['Amount'], ax=ax[0], bins=200, color='r')
ax[0].set_title('Distribution of Transaction Amount', fontsize=14)
ax[0].set_xlim([min(df['Amount']), 2000])

sns.distplot(df['Time'], ax=ax[1], color='b')
ax[1].set_title('Distribution of Transaction Time', fontsize=14)
ax[1].set_xlim([min(df['Time']), max(df['Time'])])
```

```python
corr = df.corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, cmap='coolwarm_r')
```

```python
colors = ['#B3F9C5', '#f9c5b3']

fig, ax = plt.subplots(2,4, figsize=(20,10))

sns.boxplot(x='Class', y='V17', data=df, palette=colors, ax=ax[0,0])
ax[0,0].set_title('V17 vs Class Negative Correlation', fontsize=12)
ax[0,0].grid(True)
ax[0,0].annotate('Annotate_example', xy=(1,-25), xytext=(0.01,-20), 
                 arrowprops=dict(facecolor='red'), fontsize=12)

sns.boxplot(x='Class', y='V14', data=df, palette=colors, ax=ax[0,1])
ax[0,1].set_title('V14 vs Class Negative Correlation', fontsize=12)

sns.boxplot(x='Class', y='V12', data=df, palette=colors, ax=ax[0,2])
ax[0,2].set_title('V12 vs Class Negative Correlation', fontsize=12)

sns.boxplot(x='Class', y='V10', data=df, palette=colors, ax=ax[0,3])
ax[0,3].set_title('V10 vs Class Negative Correlation', fontsize=12)

sns.boxplot(x='Class', y='V4', data=df, palette=colors, ax=ax[1,0])
ax[1,0].set_title('V4 vs Class Negative Correlation', fontsize=12)

sns.boxplot(x='Class', y='V4', data=df, palette=colors, ax=ax[1,1])
ax[1,1].set_title('V4 vs Class Negative Correlation', fontsize=12)

sns.boxplot(x='Class', y='V2', data=df, palette=colors, ax=ax[1,2])
ax[1,2].set_title('V2 vs Class Negative Correlation', fontsize=12)

sns.boxplot(x='Class', y='V19', data=df, palette=colors, ax=ax[1,3])
ax[1,3].set_title('V19 vs Class Negative Correlation', fontsize=12)

```
