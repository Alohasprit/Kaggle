```python
import pandas as pd
```

```python
train = pd.read_csv('train_1006.csv')
test = pd.read_csv('test_1006.csv')
```

```python
train = train.drop('fullVisitorId', axis=1)
x_train=train.drop('totals.transactionRevenue', axis=1)
y_train=train['totals.transactionRevenue']
```

```python
x_test=test.drop('fullVisitorId', axis=1)
id_test=test['fullVisitorId']
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(random_state=100)
model
```

```python
model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
print(predictions.shape)
predictions=pd.Series(predictions, name='totals.transactionRevenue')
```

```python
result = pd.concat([id_test,predictions], axis=1)
```

```python
submission=pd.read_csv('../data/sample_submission.csv')
```

```python
submission.shape
```

```python
submission.head()
```

```python
result = result.groupby('fullVisitorId').sum()
```

```python
submission = submission.merge(result, how='left', on='fullVisitorId')
```

```python
submission['totals.transactionRevenue']= submission['totals.transactionRevenue'].fillna(0)
```
