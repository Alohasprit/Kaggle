## 1. Preprocessing 

### 1-1. null data
null 값을 보면 컬럼별로 동일한 null 값을 가지는 컬럼 패턴이 보임.  
해당 값은 실수로 누락되었다기 보다는 실제 데이터가 없는 가능성이 높아서 null 값을 0으로 치환하는 것이 바람직해 보임

```python
null_df=df.isnull().sum().reset_index().rename(columns={'index':'col_name', 0:'num_null'}).sort_values(by='num_null', ascending=False)
null_df['percent_of_num_null']=null_df['num_null']/df.shape[0]*100
null_df.reset_index(inplace=True)

null_df
```

```python
# C50, C18, C78, Y 컬럼의 null 값을 비교하면 동일하게 null 값을 보임
df.loc[df['C50'].isnull(),['C50','C18','C78','Y']]
```

```python
# C1 ~ C117 컬럼 null 값을 0으로 채우기
for i in range(117):
    df['C'+str(i+1)]=df['C'+str(i+1)].fillna(0)
```

```python
df.head()
```

### 1-2 datetime data

```python
df['year']=df['END_TM'].apply(lambda x: x[:4])
df['month']=df['END_TM'].apply(lambda x: x[5:7])
df['days']=df['END_TM'].apply(lambda x: x[8:10])
df['time']=df['END_TM'].apply(lambda x: x[-8:])
df['hour']=df['END_TM'].apply(lambda x: x[-8:-6])
df['END_TM']=df['END_TM'].apply(lambda x: x[:10])
df['weekday']=pd.to_datetime(df['END_TM']).dt.dayofweek

df.head()
```

### 1-3 Categoracal data encoding

```python
df[['A','B']].isnull().sum()
```
